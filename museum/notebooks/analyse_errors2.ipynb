```python
for c in df.columns:
    df[f"missing_{c}"] = pd.to_numeric(df.isnull()[c])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr = df.corr(numeric_only=True)

# Create a mask for perfect correlations (1 or -1)
# We keep the lower triangle and remove perfect correlations
mask = np.triu(np.ones_like(corr, dtype=bool))  # Create upper triangle mask
mask = mask | (np.abs(corr) > 0.9999)  # Add perfect correlation mask

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Create heatmap with customizations
sns.heatmap(corr,
            fmt='.2f',
            mask=mask,
            cmap='coolwarm',
            cbar_kws={'label': 'Correlation'},
            square=True,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(rotation=0, fontsize=8)

# Adjust annotation text size
plt.gca().set_xticklabels(plt.gca().get_xticklabels(), fontsize=8)
plt.gca().set_yticklabels(plt.gca().get_yticklabels(), fontsize=8)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plotimport seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Calculate correlation matrix
corr = df.corr(numeric_only=True)

# Create a mask for perfect correlations (1 or -1)
mask = np.triu(np.ones_like(corr, dtype=bool))
mask = mask | (np.abs(corr) > 0.9999) 
mask = mask | (np.abs(corr) == 0)

# Set up the matplotlib figure with larger size
plt.figure(figsize=(16, 14))

# Create heatmap with customizations

# Rotate x-axis labels for better readability and make them smaller
plt.xticks(rotation=90, ha='right', fontsize=6)
plt.yticks(rotation=0, fontsize=6)

# Increase spacing for labels
plt.tight_layout()

# Add more padding at the bottom for x-axis labels
plt.subplots_adjust(bottom=0.2)

plt.show()
plt.show()
```
