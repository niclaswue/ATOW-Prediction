```python
import pandas as pd
import numpy as np

def calculate_typical_values(route, aircraft_type):
    data = pd.read
    # Filter data for the specific route and aircraft type
    filtered_data = data[(data['origin'] == route[0]) & 
                         (data['destination'] == route[1]) & 
                         (data['aircraft_type'] == aircraft_type)]
    
    # Function to remove outliers using IQR method
    def remove_outliers(series):
        Q1 = series.quantile(0.25)
        Q3 = series.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        return series[(series >= lower_bound) & (series <= upper_bound)]
    
    # Calculate typical load factor
    load_factors = remove_outliers(filtered_data['load_factor'])
    typical_load_factor = load_factors.mean()
    
    # Calculate typical payload
    payloads = remove_outliers(filtered_data['payload'])
    typical_payload = payloads.mean()
    
    return typical_load_factor, typical_payload


typical_load_factor, typical_payload = calculate_typical_values(challenge_data, route, aircraft_type)

print(f"Typical Load Factor: {typical_load_factor:.2f}")
print(f"Typical Payload: {typical_payload:.2f}")
```
